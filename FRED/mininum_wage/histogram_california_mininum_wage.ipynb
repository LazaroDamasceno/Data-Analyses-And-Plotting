```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('California_Mininum_Wage_Rate.csv')
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.histplot(data=df, legend=False)

ax.set_ylabel('Plateau')
ax.set_title('California Mininum Wage', fontweight='bold')
sns.despine()

plt.tight_layout()
plt.show()
```
