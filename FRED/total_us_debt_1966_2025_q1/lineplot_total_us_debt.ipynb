```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('GFDEBTN.csv')
```

```python
df_melt = df.melt(
    id_vars='observation_date',
    value_vars='GFDEBTN',
    var_name='growth'
)

df_melt = df_melt.drop(columns=['growth'])
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.lineplot(data=df_melt, y='value', x='observation_date')

sns.despine()
ax.xaxis.set_visible(False)
ax.set_ylabel('US Federal Debt Growth (In Millions)')
ax.set_title('Federal Debt Growth From From 1966 To 2025 (Q1)', fontweight='bold')

plt.tight_layout()
plt.show()
```
