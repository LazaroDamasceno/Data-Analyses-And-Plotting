```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('GFDEBTN.csv')
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.lineplot(data=df, y='GFDEBTN', x='observation_date')

sns.despine()
ax.xaxis.set_visible(False)
ax.set_ylabel('US Debt Growth (USD)')
ax.set_title('US Debt Growth From From 1966 To 2025 (Q1)', fontweight='bold')

plt.tight_layout()
plt.show()
```
