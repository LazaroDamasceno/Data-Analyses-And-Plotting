```python
import sidrapy
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = sidrapy.get_table('4714', 3, 'all')
```

```python
df.head()
```

```python
df = df.iloc[1:, :].reset_index(drop=True)
```

```python
cols = ['D3N', 'D1N', 'V']
```

```python
df = df[cols]
```

```python
df.columns
```

```python
df.dtypes
```

```python
df['V'] = df['V'].astype('float')
```

```python
df.dtypes
```

```python
df = df.rename(columns={
    'D3N': 'tipo',
    'D1N': 'uf',
    'V': 'valor'
})
```

```python
df.columns
```

```python
df.head()
```

```python
df['tipo'].unique()
```

```python
mapa = gpd.read_file('BR_UF_2024/')
```

```python
mapa
```

```python
df_pop = df.loc[df['tipo'] == 'População residente'].reset_index(drop=True)
```

```python
df_pop.head()
```

```python
mapa_pop = gpd.GeoDataFrame(pd.merge(left=df_pop.drop('tipo', axis=1), right=mapa, left_on='uf', right_on='NM_UF'))
```

```python
mapa_pop.head()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

mapa_pop.plot(
    column='valor',
    legend=True,
    ax=ax
)

ax.set_title('População brasileira nas UF (2022)')

plt.tight_layout()
plt.show()
```

```python
df_area_uf = df.loc[df['tipo'] == 'Área da unidade territorial'].reset_index(drop=True)
```

```python
df_area_uf.head()
```

```python
mapa_area_uf = gpd.GeoDataFrame(pd.merge(left=df_area_uf.drop('tipo', axis=1), right=mapa, left_on='uf', right_on='NM_UF'))
```

```python
mapa_area_uf.head()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

mapa_area_uf.plot(
    column='valor',
    legend=True,
    ax=ax
)

ax.set_title('Área territorial das UF em km2 (2022)')

plt.tight_layout()
plt.show()
```

```python
df_dens = df.loc[df['tipo'] == 'Densidade demográfica'].reset_index(drop=True)
```

```python
df_dens.head()
```

```python
mapa_dens = gpd.GeoDataFrame(pd.merge(left=df_dens.drop('tipo', axis=1), right=mapa, left_on='uf', right_on='NM_UF'))
```

```python
mapa_dens.head()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

mapa_dens.plot(
    column='valor',
    legend=True,
    ax=ax
)

ax.set_title('Densidade demográfica das UF por habitante por km2 (2022)')

plt.tight_layout()
plt.show()
```
