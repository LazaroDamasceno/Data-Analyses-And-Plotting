```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
gdf = gpd.read_file('us_obesity_by_state/')
```

```python
gdf.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

gdf.plot(
    column='Obesity',
    legend=True,
    ax=ax
)

plt.tight_layout()
fig.savefig('ChoroplethMap_UsObesityByStates_2015.png', dpi=300, bbox_inches='tight')
plt.show()
```
