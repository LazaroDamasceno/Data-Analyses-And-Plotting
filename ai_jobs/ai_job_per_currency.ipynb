```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("ai_job_dataset.csv")
df_cleaned = df.dropna()
```

```python
df_grouped = df_cleaned.groupby('salary_currency')['job_title']
df_grouped = df_grouped.size()
df_grouped = df_grouped.reset_index(name="value")
```

```python
df_sorted = df_grouped.sort_values(by=('value'), ascending=False)
df_sorted = df_sorted.reset_index()
df_sorted = df_sorted.drop(columns=['index'])
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.barplot(data=df_sorted, orient='h', ax=ax, x='value', y='salary_currency')

ax.set_title("AI Jobs By Currency", fontweight="bold")
ax.set_ylabel("")
ax.xaxis.set_visible(False)
ax.spines[['top', 'right', 'bottom']].set_visible(False)

plt.tight_layout()
plt.show()
```
