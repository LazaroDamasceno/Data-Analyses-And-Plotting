```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.histplot(data=df_sorted, x='value')

ax.set_title("", fontweight="bold")
ax.set_xlabel("Average Salary")
ax.set_ylabel("")
sns.despine()
ax.legend(title="")

plt.tight_layout()
plt.show()
```
