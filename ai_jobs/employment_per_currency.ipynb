```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ai_job_dataset.csv")
df_cleaned = df.dropna()
```

```python
df_grouped = df_cleaned.groupby(['employment_type', 'salary_currency'])
df_grouped = df_grouped.size()
df_grouped = df_grouped.reset_index(name="value")
df_grouped = df_grouped.sort_values(by=("salary_currency"), ascending=False)
df_grouped = df_grouped.reset_index(drop=True)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.barplot(
    data=df_grouped,
    x='value',
    y='salary_currency',
    hue='employment_type',
    ax=ax,
    orient="h"
)

for container in ax.containers:
    ax.bar_label(container, fmt="%.0f")

ax.set_title("Type Of Employment By Currency", fontweight="bold")
ax.set_ylabel("")
ax.xaxis.set_visible(False)
ax.spines[['top', 'right', 'bottom']].set_visible(False)
ax.legend(title="Types")

plt.tight_layout()
plt.show()
```
