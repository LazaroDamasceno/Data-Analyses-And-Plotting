```python
import kagglehub
from kagglehub import KaggleDatasetAdapter 
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "ai_job_dataset.csv"
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "bismasajjad/global-ai-job-market-and-salary-trends-2025",
  file_path,
)
```

```python
df_grouped = df.groupby('salary_currency')['job_title'].size().reset_index(name="value")
```

```python
df_sorted = df_grouped.sort_values(by=('value'), ascending=False).reset_index(drop=True)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.barplot(data=df_sorted, orient='h', ax=ax, x='value', y='salary_currency')

ax.bar_label(ax.containers[0])

ax.set_title("AI Jobs By Currency", fontweight="bold")
ax.set_ylabel('')
ax.xaxis.set_visible(False)
sns.despine(bottom=True)

plt.tight_layout()
plt.show()
```
