```python
import kagglehub
from kagglehub import KaggleDatasetAdapter 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "ai_job_dataset.csv"
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "bismasajjad/global-ai-job-market-and-salary-trends-2025",
  file_path,
)
df_cleaned = df.dropna()
```

```python
df_grouped = df_cleaned.groupby(['company_location', 'company_size']).size().reset_index(name="value")
```

```python
fig, ax = plt.subplots(figsize=(12, 24))

sns.barplot(
    data=df_grouped,
    x='value',
    y='company_location',
    hue='company_size',
    ax=ax,
    orient='h'
)

for container in ax.containers:
    ax.bar_label(container, fmt="%.0f")

ax.set_title("Company Size By Country", fontweight="bold")
ax.set_ylabel("")
ax.xaxis.set_visible(False)
sns.despine(bottom=True)
ax.legend(title="Sizes", loc="lower left", bbox_to_anchor=(0, 1.02, 1, 0.2), mode="expand", borderaxespad=0, ncol=3)

plt.tight_layout()
plt.show()
```
