```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("ai_job_dataset.csv")
```

```python
df_cleaned = df.dropna()
```

```python
df_longer = df_cleaned.melt(
    id_vars="salary_currency",
    value_vars="job_title",
    var_name="jobs"
)
```

```python
df_pivot = df_longer.pivot_table(
    index="salary_currency",
    columns="jobs",
    aggfunc="count"
)
```

```python
df_stacked = df_pivot.stack().reset_index()
```

```python
df_sorted = df_stacked.sort_values(by=('value'), ascending=False)
df_sorted = df_sorted.reset_index()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.barplot(
    data=df_sorted,
    ax=ax,
    x='value',
    y='salary_currency',
    orient='h'
)

ax.set_title("AI jobs by currency", fontweight="bold")
ax.bar_label(ax.containers[0])
ax.set_ylabel("")
ax.spines[['top', 'right', 'bottom']].set_visible(False)
ax.xaxis.set_visible(False)

plt.tight_layout()
plt.show()
```
