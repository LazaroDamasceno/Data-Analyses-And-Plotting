```python
import seaborn as sns
import matplotlib.pyplot as plt
import kagglehub
from kagglehub import KaggleDatasetAdapter
```

```python
file_path = "ai_job_dataset.csv"
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "bismasajjad/global-ai-job-market-and-salary-trends-2025",
  file_path,
)
df_cleaned = df.dropna()
```

```python
filter_2024 = df_cleaned['posting_date'].str.contains('2024')
df_2024_filtered = df_cleaned[filter_2024]
```

```python
df_2024_grouped = df_2024_filtered.groupby('posting_date')['job_title']
df_2024_grouped = df_2024_grouped.size()
df_2024_grouped = df_2024_grouped.reset_index(name="value")
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.lineplot(data=df_2024_grouped, x='posting_date', y='value')

ax.xaxis.set_visible(False)
sns.despine(bottom=True)
ax.set_title("AI Jobs In 2024 Globally", fontweight="bold")

plt.tight_layout()
plt.show()
```
