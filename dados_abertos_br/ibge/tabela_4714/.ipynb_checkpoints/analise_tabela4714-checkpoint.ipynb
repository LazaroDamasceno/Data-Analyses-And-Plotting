```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
```

# 1 CRIAÇÃO DO DATAFRAME

```python
df = pd.read_csv('tabela_4714_limpa.csv')
```

```python
df.head()
```

# 2 SELEÇÃO DE COLUNAS

```python
df['variavel'].value_counts()
```

```python
df = df.rename(columns={
    'uf': 'NM_UF'
})
```

```python
def transformar_df(df, variavel):
    df = df.loc[df['variavel'] == variavel].reset_index(drop=True)
    df = df.iloc[:, range(2)]
    return df
```

# 3 TRANSFORMAÇÃO DE COLUNAS

# 3.1 TRANSFORMAÇÃO DE COLUNAS: POPULAÇÃO DAS UF

```python
df_populacao = transformar_df(df, 'População residente')
```

```python
df_populacao
```

```python
df_populacao = df_populacao.rename(columns={
    'valor': 'populacao_uf'
})
```

```python
df_populacao.columns
```

```python
total = df_populacao['populacao_uf'].sum()

df_populacao['populacao_uf'] = df_populacao['populacao_uf'].apply(lambda x: (x / total) * 100.0).round(2)

df_populacao['populacao_uf']
```

# 3.2 TRANSFORMAÇÃO DE COLUNAS: ÁREA TERRITÓRIAL DA UF

```python
df_area_uf = transformar_df(df, 'Área da unidade territorial')
```

```python
df_area_uf
```

```python
df_area_uf = df_area_uf.rename(columns={
    'valor': 'area_uf'
})
```

```python
df_area_uf.columns
```

```python
total = df_area_uf['area_uf'].sum()

df_area_uf['area_uf'] =df_area_uf['area_uf'].apply(lambda x: (x / total) * 100.0).round(2)

df_area_uf['area_uf']
```

# 3.3 TRANSFORMAÇÃO DE COLUNAS: DENSIDADE POPULACIONAL

```python
df_densidade = transformar_df(df, 'Densidade demográfica')
```

```python
df_densidade
```

```python
df_densidade = df_densidade.rename(columns={
    'valor': 'densidade_demografica'
})
```

```python
df_densidade.columns
```

# 4. CRIAÇÃO DO MAPA COROPLÉTICO

```python
mapa = gpd.read_file('BR_UF_2024/')
```

```python
mapa
```

```python
mapa = mapa.merge(right=df_area_uf, on='NM_UF').merge(right=df_densidade, on='NM_UF').merge(right=df_populacao, on='NM_UF')
```

```python
mapa
```

# 5 MAPAS COROPLÉTICOS

# 5.1 MAPAS COROPLÉTICOS: ÁREA TERRITÓRIAL DAS UF

```python
fig, ax = plt.subplots(figsize=(10, 6))

mapa.plot(
    column='area_uf',
    legend=True,
    legend_kwds={
        'label': '(%)'
    },
    ax=ax
)

ax.set_title('Área territórial das UF', fontweight='bold')

plt.tight_layout()
fig.savefig('MapaCoropletico_AreaTerritoralUF.png', dpi=300, bbox_inches='tight')
plt.show()
```

# 5.2 MAPAS COROPLÉTICOS: DENSIDADE DEMOGRÁFICA

```python
fig, ax = plt.subplots(figsize=(10, 6))

mapa.plot(
    column='densidade_demografica',
    legend=True,
    ax=ax
)

ax.set_title('Densidade demográfica das UF', fontweight='bold')

plt.tight_layout()
fig.savefig('MapaCoropletico_DensidadeDemograficaUF.png', dpi=300, bbox_inches='tight')
plt.show()
```

# 5.2 MAPAS COROPLÉTICOS: POPULAÇÃO DAS UF

```python
fig, ax = plt.subplots(figsize=(10, 6))

mapa.plot(
    column='populacao_uf',
    legend=True,
    legend_kwds={
        'label': '(%)'
    },
    ax=ax
)

ax.set_title('Populacação das UF', fontweight='bold')

plt.tight_layout()
fig.savefig('MapaCoropletico_PopulacaoUF.png', dpi=300, bbox_inches='tight')
plt.show()
```
