```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import seaborn as sns
```

```python
df = pd.read_csv(
    "https://ourworldindata.org/grapher/electoral-democracy-index.csv?v=1&csvType=full&useColumnShortNames=true", 
    storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'}
)
```

```python
df.head()
```

```python
df = df.drop(columns={
    'Code', 'owid_region'
})
```

```python
df.columns
```

```python
df = df.rename(columns={
    'electdem_vdem__estimate_best': 'electoral_democracy_index',
})
```

```python
old_titles = df.columns.unique()
new_titles = [title.lower() for title in old_titles]
renamed_titles = dict(zip(old_titles, new_titles))
```

```python
df = df.rename(columns=renamed_titles)
```

```python
df.columns
```

```python
fig = plt.figure(figsize=(12, 8))

gs = GridSpec(2, 2, figure=fig)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[1, 0])
ax3 = fig.add_subplot(gs[:, 1])

sns.lineplot(df, x='year', y='electoral_democracy_index', ax=ax1)
sns.histplot(df, x='electoral_democracy_index', ax=ax2)
sns.boxplot(df, x='electoral_democracy_index', ax=ax3)

ax1.set_ylabel('Electoral Democracy Index')

axes = [ax1, ax2, ax3]
titles = ['Line Diagram', 'Histogram', 'Boxplot']
x_labels = ['Year', 'Electoral Democracy Index', 'Electoral Democracy Index']
for ax, x_label, title in zip(axes,  x_labels, titles):
    ax.set_xlabel(x_label)
    ax.set_title(title)

plt.suptitle("World's Electoral Democracy Index", fontweight='bold')
plt.tight_layout()
fig.savefig('World_Electoral_Democracy_world.png', dpi=300, bbox_inches='tight')
plt.show()
```
