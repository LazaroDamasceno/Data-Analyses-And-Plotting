```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
gdf = gpd.read_file('us_geo_data/')
```

```python
gdf.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

gdf.plot(
    column='Obesity',
    ax=ax,
    legend=True
)

ax.set_title('Obesity by US State in 2015', fontweight='bold')

plt.tight_layout()
fig.savefig('obesity_by_us_state_2015.png', dpi=300, bbox_inches='tight')
plt.show()
```
